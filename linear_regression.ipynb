```python
#Linear regression example
# This is a very simple example of using two scipy tools
# for linear regression, polyfit and stats.linregress

# import statements
from scipy import linspace, polyval, polyfit, sqrt, stats, randn
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
#Sample data creation
#number of points
n=50
t=linspace(-5,5,n)
```

```python
#parameters
a=0.8; b=-4
x=polyval([a,b],t)
```

```python
#add some noise
xn=x+randn(n)
```

```python
#Linear regressison -polyfit - polyfit can be used other orders polys
(ar,br)=polyfit(t,xn,1)
xr=polyval([ar,br],t)
```

```python
#compute the mean square error
err=sqrt(sum((xr-xn)**2)/n)
```

```python
print('Linear regression using polyfit')
print('parameters: a=%.2f b=%.2f \nregression: a=%.2f b=%.2f, ms error= %.3f' % (a,b,ar,br,err))
```

```python
#matplotlib ploting
plt.title('Linear Regression Example using polyfit')
plt.plot(t,x,'g.--')
plt.plot(t,xn,'k.')
plt.plot(t,xr,'r.-')
plt.legend(['original','plus noise', 'regression'])

plt.show()
```

```python
#Linear regression using stats.linregress
(a_s,b_s,r,tt,stderr)=stats.linregress(t,xn)
x_s=polyval([a_s,b_s],t)
print('Linear regression using stats.linregress')
print('parameters: a=%.2f b=%.2f \nregression: a=%.2f b=%.2f, std error= %.3f' % (a,b,a_s,b_s,stderr))
```

```python
#matplotlib ploting
plt.title('Linear Regression Example using stats.linregress')
plt.plot(t,x,'g.--')
plt.plot(t,xn,'k.')
plt.plot(t,x_s,'r.-')
plt.legend(['original','plus noise', 'regression'])

plt.show()
```

#### Note
Standard error is less if we use **stats.linregress** compared to **polyfit**
